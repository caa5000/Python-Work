#Problem 1(a).  A collectible card class.

You are tasked with organizing information about a Magic: The Gathering card collection.  You are particularly interested in which cards are most rare, and therefore most valuable.

The MTG JSON website (http://mtgjson.com/) has information about different sets of Magic: The Gathering cards in JSON format.  Using the [Example Card](http://mtgjson.com/#exampleCard) information, create a MagicCard class to store relevant information about Magic cards in Python.

Consider:

* How should someone initialize an instance of your class?
* What functions should your class have?

```python
#Looking for different factors of each card to determine its rarity. 
#will need to include basic identification factors. 
#name, rarity, id, colors, types (i.e. creature, land etc)

#How should someone initialize an instance of my class?
    #Perhaps the class can be initiated by 'searching' for any of the five possibilities mentioned above.
    #If it's initiatlized with nothing, sort by rarity for default?
    
#What functions should your class have? Since we are querying information I don't think we'll need initiation.
    #getName, getRarity, getID, getColors, getTypes
    
class MagicCard(object):
    
    def __init__(self, name="", rarity="", ID="", colors="", types=""):
        self.name = name
        self.rarity = rarity
        self.ID = ID
        self.colors = colors
        self.types = types
    
    def get_name(self):
        return self.name
    
    def get_rarity(self):
        return self.rarity
    
    def get_ID(self):
        return self.ID
    
    def get_colors(self):
        return self.colors
    
    def get_types(self):
        return self.types
```

Download the data for the Dragons of Tarkir set from the [Individual Sets](http://mtgjson.com/#individualSets) section of the MTG JSON page.  Create a MagicCardSet class to store relevant information about a set of Magic cards, and create an instance of the class which contains all of the information about the Dragons of Tarkir set.

#Problem 1(c).  Dragons of Tarkir.

What are the rarest cards in the Dragons of Tarkir set?  What colors are they associated with?

Write Python code to answer this question, then create a markdown cell which explains your method and conclusions.

```python
#Read in Dragons of Tarkir set information
with open('DTK.json', encoding="utf8") as DTK:
    dragonsTarkir = json.loads(DTK.read())
dragonsTarkir
```

```python
#Assign values to run through class. Thnaks to classmates for help with 'cards.get'

card_list = [MagicCard(cards["name"], cards["rarity"], cards["id"], 
                       cards.get("colors", ""), cards["type"]) for cards 
             in dragonsTarkir["cards"]]
card_list
```

```python
import pandas as pd
```

```python
#tarkir_list
```

```python
#What are the rarest Cards in the Dragons of Tarkir set? What colors are they associated with?

column_names = ("Name", "Rarity", "ID", "Colors", "Type")
tarkir_list = []
for cards in card_list:
    #has_instagram = "instagram" in photo.get_tags()
    tarkir_list.append(dict(zip(column_names, [cards.get_name(), cards.get_rarity(), 
                                               cards.get_ID(), cards.get_colors(), 
                                               cards.get_types()])))
tarkir_df = pd.DataFrame(tarkir_list)
tarkir_df
```

```python
tarkir_list
```

```python
#v1. "Mythic Rare" is the rarest type of magic card via google search.
#v2. After doing some research I found that the two most valuable cards are 'rare', so I added them back in! 

#Create a dataframe with only the mythic rares.
mythic_rare_df = pd.DataFrame()
mythic_rare_df = tarkir_df[(tarkir_df.Rarity == "Mythic Rare") | (tarkir_df.Rarity == "Rare")]
mythic_rare_df

#df2 = df[(df.a != -1) | (df.b != -1)]

```

```python
#v1. Run a check for the colors of the mythic rares.
#v2. Added 'rare' back in! 
mythic_rare_df['Colors'].apply(pd.value_counts).count()

```

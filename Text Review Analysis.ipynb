#Problem 1(b).  Analyzing review text.

Add columns to your DataFrame for the length of a review, the number of exclamation points in a review, and the fraction of people who rated a review helpful.  If no people voted on whether a problem was helpful, the corresponding entry in your percentage helpful column should be ```NaN```.

````python
#Create function to count length of review. 

def lengthreview(row):
    return (len(row['Text']))

#Create function to count number of exclamation points.

def exclamationpoints(row):
    return (row['Text'].count('!'))

#Create function to return fraction of people who thought review was helpful. If no one voted (Rated Reviews) then make null.

def fractionhelpful(row):
    return (float(row['Helpful'])/float(row['Rated Reviews']))
    
    
````

````python
#coerce numbers to numeric. 
#pd.to_numeric('Rated Reviews', errors='coerce')
#pd.to_numeric('Helpful', errors='coerce')
#pd.to_numeric('Review Length', errors='coerce')


#Set any zeroes to NaN for calculation purposes.
review_df=review_df.replace('0', np.nan)

review_df
````

````python

#add column for length of review.
review_df['Review Length'] = review_df.apply(lengthreview, axis=1)
#add column for number of exclamation points.
review_df['Exclamation Points'] = review_df.apply(exclamationpoints, axis=1)

#add column for fraction of people who rated a review helpful. If no one voted put in as NaN. 
review_df['Helpful Fraction'] = review_df.apply(fractionhelpful, axis=1)

review_df=review_df.replace('0', np.nan)

review_df
````
